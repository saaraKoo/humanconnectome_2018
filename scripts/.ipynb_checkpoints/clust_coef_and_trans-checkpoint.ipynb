```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update(mpl.rcParamsDefault)
import numpy as np
import networkx as nx
import igraph as igp
```

```python
# Load the data
path = '/anaconda/envs/testenv_notebooks/networks/graphs/'
sub = os.listdir(path)
G = {}
```

```python
# Generate networkx graphs from the adjacency matrices
for i in range(0, len(sub)):
    with open(path + sub[i]) as adj_csv:
        adj = np.loadtxt(adj_csv, delimiter=',')
        G['{}'.format(sub[i][:-11])] = nx.from_numpy_array(adj)
```

```python
# Clustering coefficients for each node of each subject
clustering_coefficients = [nx.clustering(G[key]) for key in G]
```

```python
# Average clustering coefficients for each subject
avg_clustering_coefficients = [nx.average_clustering(G[key]) for key in G]
```

```python
# Path for plots
path = '/anaconda/envs/testenv_notebooks/networks/figs/'
```

```python
# Plot histogram of average clustering coefficients
mean = np.mean(avg_clustering_coefficients)

plt.hist(avg_clustering_coefficients)
plt.xlabel('Clustering coefficient')
plt.ylabel('Frequence')
plt.title('Histogram of average clustering coefficients')
plt.axvline(mean, c="r", ls="--", label="Mean")

plt.grid(color='lightgrey', ls='dashed')
plt.savefig(path + 'ACC_hist.pdf', format='pdf')
plt.show()
```

```python
# Plot cumulative distribution function for average clustering coefficients
N = len(avg_clustering_coefficients)
X = np.sort(avg_clustering_coefficients)
F = np.array(range(N))/float(N)

plt.plot(X, F)
plt.xlabel('Average clustering coefficient')
plt.ylabel(r'$P(X\leq c)$')
plt.title('CDF of average clustering coefficient')

plt.grid(color='lightgrey', ls='dashed')
plt.savefig(path + 'ACC_CDF.pdf', format='pdf')
plt.show()
```

```python
# Transitivity
trans = [nx.transitivity(G[key]) for key in G]
```

```python
# Plot histogram of transitivities
mean = np.mean(trans)

plt.hist(trans)
plt.xlabel('Transitivity')
plt.ylabel('Frequence')
plt.title('Histogram of average transitivity')
plt.axvline(mean, c="r", ls="--", label="Mean")

plt.grid(color='lightgrey', ls='dashed')
plt.savefig(path + 'trans_hist.pdf', format='pdf')
plt.show()
```

```python
# Plot cumulative distribution function for transitivity
N = len(trans)
X = np.sort(trans)
F = np.array(range(N))/float(N)

plt.plot(X, F)
plt.xlabel('Transitivity')
plt.ylabel(r'$P(X\leq T)$')
plt.title('CDF of average transitivity')

plt.grid(color='lightgrey', ls='dashed')
plt.savefig(path + 'trans_CDF.pdf', format='pdf')
plt.show()
```
